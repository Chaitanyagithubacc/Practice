```python
import pandas as pd

# Sample DataFrame for demonstration
table = pd.DataFrame({
    'Order Date': ['2016-11-07', '2016-11-08', '2015-10-18'],
    'Ship Date': ['2015-10-17', '2015-10-18', '2015-10-19'],
    'Ship Mode': ['Standard Class', 'Second Class', 'Standard Class']
})

# Convert to datetime
table['Order Date'] = pd.to_datetime(table['Order Date'])
table['Ship Date'] = pd.to_datetime(table['Ship Date'])

# Filter DataFrame
result = table[(table['Order Date'] >= '2016-11-08') & 
               (table['Ship Date'] <= '2015-10-18') & 
               (table['Ship Mode'].isin(['Standard Class', 'Second Class']))]['Order Date']

print(result)
```
