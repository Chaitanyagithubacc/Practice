```python
import pandas as pd

# Assuming 'House' is your DataFrame
null_counts = categorical1.isnull().sum().sort_values(ascending=False)
null_counts_percentage = categorical1.isnull().sum()/len(categorical1)*100
data_types = categorical1.dtypes

# Combine the two into a single DataFrame and align by the sorted null counts index
combined_info = pd.DataFrame({
    'Null Count': null_counts,
    'Null counts%' : null_counts_percentage,
    'Data Type': data_types[null_counts_percentage.index]  # Align with the sorted null counts
})

# Sort by 'Null counts%' in descending order
combined_info = combined_info.sort_values(by='Null counts%', ascending=False)

# Display the combined DataFrame
print(combined_info)
```
